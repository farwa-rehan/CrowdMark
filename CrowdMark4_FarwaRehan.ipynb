```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point, tolerance = 1e-10, max_steps = 10000):

    # estimate derivative using symmetric difference quotient
    def deriv(f, base_point):
        return (f(base_point + 10**(-10)) - f(base_point - 10**(-10))) / (2 * 10**(-10))

    x = initial_point
    x_coords = [initial_point]
    y_coords = [f(initial_point)]

    for n in range(max_steps):
        slope = deriv(f, x)

        if abs(slope) < tolerance:
            break

        x = x - learning_rate * slope

        x_coords.append(x)
        y_coords.append(f(x))

    # plotting portion
    plot_range = np.linspace(min(x_coords)-0.5, max(x_coords)+0.5, 10000)
    function_range = [f(i) for i in plot_range]

    plt.plot(plot_range, function_range, label = "f(x)")
    plt.plot(x_coords, y_coords, marker = "o", label = "descent path")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title("Gradient Descent for f(x)")
    plt.legend()
    plt.grid()
    plt.show()

    print("Minimum coordinate:", (round(x_coords[-1],3), round(y_coords[-1],3)))

    return round(x_coords[-1],3), round(y_coords[-1],3)
```

```python
#example 1: f(x) = x^2
def f(x):
    return x**2
# showcases a nice stable descent to 0
gradient_descent(f, learning_rate = 0.5, initial_point = 2)
```

```python
# example 2: f(x) = x^4 -2x^2
def f1(x):
    return x**4 - 2*x**2

gradient_descent(f1, learning_rate = 0.1, initial_point = 2) #finding the minimum at positive 1
gradient_descent(f1, learning_rate = 0.1, initial_point = -2) #finding maximum at negative 1
gradient_descent(f1, learning_rate = 0.1, initial_point = 0) # showng gradient descent stays at the maximum x = 0
```

```python
#example 3: funny function (x^x)
def funny_function(x):
    if x > 0:
        return x**x
    elif x == 0:
        return 1
    else:
        return abs(x)**abs(x)

# using a small learning rate so that graph converges to 0
gradient_descent(funny_function, learning_rate = 0.005, initial_point = 0.8)
```

```python
#example 4: f(x)=|x|
def abs_function(x):
    return abs(x)

# non differentiable so gradient will bounce around 0
gradient_descent(abs_function, learning_rate = 0.2, initial_point = 2)
```
